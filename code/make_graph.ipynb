```python
import pandas as pd
```

```python
data = pd.read_csv('./crawling.csv')
```

```python
data['dateonly'] = data['datetime'].apply(lambda x: x[:10])
```

```python
data.groupby(['dateonly']).size().reset_index(name='counts').to_csv('./stat_perday.csv',index=False)
```

```python
data['monthonly'] = data['datetime'].apply(lambda x: x[:7])
```

```python
data.groupby(['monthonly']).size().reset_index(name='counts').to_csv('./stat_permonth.csv',index=False)
```
