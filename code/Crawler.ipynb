```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import ssl
import csv
```

```python
def get_bs(url):
#     html = urlopen(url)
    context = ssl._create_unverified_context()
    html = urlopen(url, context=context)

    bs = BeautifulSoup(html.read(),"html.parser",from_encoding='utf-8')
#     bs = bs.prettify(encoding='utf-8')
    return bs
```

```python
with open('./error.csv','w',newline='\n') as error:
    errorwriter = csv.writer(error)
    with open('./crawling.csv','w',encoding='utf-8',newline='\n') as writefile:
        writer = csv.writer(writefile,quoting=csv.QUOTE_ALL)
        header = ["datetime","title","service","voc_type","voc_date","text"]
        writer.writerow(header)
        for page in range(1,1709):
            
            url = "https://www.skhystec.com/service-voc.asp?gotopage=%s&b_code=s1&s_type=&s_word=&b_sclass="%page
            print ("page %d :%s"%(page,url))
            try:
                bs = get_bs(url)
            except:
                errorwriter.writerow([url])
            else:
                items = bs.find("ul",{"class":"tableBbs"}).find_all("li")
                for item in items:
                    href = item.a['href']
                    if href=="#" and not ("class" in (item.a.attrs)):
                        sub = item.find("span",{"class":"subject"}).text.replace('"','')
                        date = item.find("span",{"class":"articledate"}).text
                        writer.writerow([date,sub,"secret","","",""])
                    if not (href=="#" or "class" in (item.a.attrs) ):
                        url = "https://www.skhystec.com/%s"%(href)
                        try:
                            bs = get_bs(url)
                        except:
                            errorwriter.writerow([href])
                        else:
                            table = bs.find_all("dl",{"class":"tableArticleInfo"})
                            dds = table[0].find_all("dd")
                            subject = dds[0].text.strip().replace('"','')
                            datetime = dds[1].text.strip()

                            a = ["","",""]
                            for idx,item in enumerate(dds[3:6]):
                                a[idx] = (item.text.strip())

                            text = table[1].find("dd").text.strip().replace('"','')
                            writer.writerow([datetime,subject,a[0],a[1],a[2],text])
```
